# 📊 Ablation Results & Comparative Analysis

This notebook performs structural data aggregation, error analysis, and comparative visualization of our System-2 AI Researcher agent across all baseline and custom fine-tuned model variants.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.evaluation.analysis import ResultsAnalyzer

print("Libraries loaded successfully!")
```

## 1. Load Experiment Runs

Using the custom `ResultsAnalyzer`, we parse all executed experiment run output JSON files.

```python
analyzer = ResultsAnalyzer(results_dir="logs/experiments")
df_runs = analyzer.load_all_runs()

if df_runs.empty:
    print("No experiment runs found yet in logs/experiments. Run scripts/run_all_experiments.py first.")
else:
    print(f"Successfully loaded {len(df_runs)} runs.")
    display(df_runs.head())
```

## 2. Structural Comparison Table

Generate a summary matrix displaying F1 Score, Citation Precision, LLM Judge Score, and average step count across model variants.

```python
if not df_runs.empty:
    df_comparison = analyzer.build_comparison_table(df_runs)
    display(df_comparison)
else:
    # Display mock comparison to demonstrate metrics visualization
    mock_data = {
        "mean_f1_score": [0.42, 0.58, 0.74, 0.86],
        "mean_citation_precision": [0.0, 0.72, 0.88, 0.95],
        "mean_overall_judge": [2.3, 3.5, 4.2, 4.8],
        "mean_step_count": [1.0, 5.2, 7.8, 6.4],
        "success_rate": [0.15, 0.75, 0.90, 0.98]
    }
    df_comparison = pd.DataFrame(mock_data, index=["exp1_vanilla_llm", "exp2_rag_baseline", "exp3_agent_base", "exp6_agent_full"])
    print("--- Mock Evaluation Comparison Table ---")
    display(df_comparison)
```

## 3. Visualize Comparative Performance

Plot bar charts comparing vanilla zero-shot direct prompting, standard RAG, agent baseline, and full fine-tuned models.

```python
plt.figure(figsize=(10, 5))
sns.set_theme(style="whitegrid")
sns.barplot(x=df_comparison.index, y="mean_f1_score", data=df_comparison, palette="Blues_d")
plt.title("Ablation Study: F1 Score Comparison Across Configurations", fontsize=14)
plt.ylabel("Mean F1 Score")
plt.xlabel("Configuration Variant")
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```
